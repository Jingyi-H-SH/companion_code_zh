## 这套演示展示什么

这一章展示一种面向治理场景的 RAG 模式：先检索相关政策参考片段，再根据清单为系统评分，最后把结果合成为报告。

## 工作流程

- 加载示例系统画像和检查清单。
- 针对隐私、可解释性和公平性分别检索最相关的治理参考资料。
- 调用 LLM 进行结构化评分并返回发现项。
- 把各部分得分汇总成最终合规报告。
- 输出 JSON、Markdown 和章节得分图。

## 核心技术概念

- 基于清单的 Prompt 设计
- 用于政策/治理审查的 RAG
- 结构化章节评分
- 基于中间审查结果自动合成报告


```python
import os
import subprocess
import sys
from getpass import getpass

if not os.environ.get("OPENAI_API_KEY"):
    os.environ["OPENAI_API_KEY"] = getpass("请输入你的 OpenAI API key（输入不可见）：")
os.environ.setdefault("OPENAI_MODEL", "gpt-4.1-mini")

subprocess.run([sys.executable, "../run_demo.py"], check=True)
```
