## 这套演示展示什么

这一章展示一种“双模型审计”模式：先由一个 LLM 生成回答，再由另一个结构化审查步骤为回答打分，最后进行群体比较。

## 工作流程

- 把每个健康问题与每个用户画像组合起来。
- 为每个问题-画像组合生成一条回答。
- 再用结构化审查步骤给共情、可及性、安全性和可操作性打分。
- 按群体汇总这些得分，并检测差异。
- 最后生成偏差审计报告。

## 核心技术概念

- LLM-as-generator（生成模型）
- LLM-as-judge（审查模型）
- 面向社会影响问题的评估工作流
- 先对回答评分，再做群体层面的汇总比较
