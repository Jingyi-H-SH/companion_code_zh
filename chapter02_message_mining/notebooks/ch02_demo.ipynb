# 居民健康留言挖掘: 快速演示

这个 Notebook 属于《人工智能与健康传播：基础与应用》的配套代码。

## 这套演示展示什么

这一章展示一个基础的 LLM 分类工作流：先整理短文本，再检索主题参考资料，然后用结构化 JSON 完成分类，并进一步生成回应建议。

## 工作流程

- 先整理居民留言文本，并生成简单的特征提示。
- 再从本地 JSON 参考资料中检索最相关的主题说明片段。
- 调用 OpenAI API，返回结构化的主题和情绪标签。
- 根据分类结果生成一条面向公众的简短回应和一条教师备注。
- 保存输出结果，并生成进度看板。

## 核心技术概念

- 面向分类任务的 Prompt 设计
- 基于 JSON schema 的结构化输出
- 使用本地参考文件实现轻量级 RAG
- 把“分析”和“后续回复生成”拆开处理


```python
import os
import subprocess
import sys
from getpass import getpass

if not os.environ.get("OPENAI_API_KEY"):
    os.environ["OPENAI_API_KEY"] = getpass("请输入你的 OpenAI API key（输入不可见）：")
os.environ.setdefault("OPENAI_MODEL", "gpt-4.1-mini")

subprocess.run([sys.executable, "../run_demo.py"], check=True)
```
