# 饮水习惯助手: 讲解版 Notebook

这个 Notebook 属于《人工智能与健康传播：基础与应用》的配套代码。

## 这套演示展示什么

这一章展示如何根据用户状态实现个性化 LLM 输出。它先预测饮水风险，再检索合适的提醒策略，最后生成定制化提醒。

## 工作流程

- 从饮水日志中构建可解释的用户状态特征。
- 让 LLM 基于已知特征估计当天未完成目标的风险。
- 从本地提醒策略手册中检索匹配的策略片段。
- 生成个性化提醒、一条微行动建议和简短解释。
- 最后评估并可视化结果。

## 核心技术概念

- 基于用户状态的 Prompt 设计
- 把预测和生成放在同一条工作流里
- 把 RAG 用于“策略选择”，而不只是长文本问答
- 面向读者的个性化生成


```python
import os
import subprocess
import sys
from getpass import getpass

if not os.environ.get("OPENAI_API_KEY"):
    os.environ["OPENAI_API_KEY"] = getpass("请输入你的 OpenAI API key（输入不可见）：")
os.environ.setdefault("OPENAI_MODEL", "gpt-4.1-mini")

subprocess.run([sys.executable, "../run_demo.py"], check=True)
```
